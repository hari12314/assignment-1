```python
conn = psycopg2.connect(
user = "postgres",
dbname = "food_data",
password = "Hari",
host = "localhost",
port = 5432)
conn.autocommit = True
cursor = conn.cursor()
```

```python
schema_sql = """
CREATE TABLE IF NOT EXISTS Providers (
    Provider_ID INTEGER PRIMARY KEY,
    Name TEXT,
    Type TEXT,
    Address TEXT,
    City TEXT,
    Contact TEXT
);

CREATE TABLE IF NOT EXISTS Receivers (
    Receiver_ID INTEGER PRIMARY KEY,
    Name TEXT,
    Type TEXT,
    City TEXT,
    Contact TEXT
);

CREATE TABLE IF NOT EXISTS Food_Listings (
    Food_ID INTEGER PRIMARY KEY,
    Food_Name TEXT,
    Quantity INTEGER,
    Expiry_Date DATE,
    Provider_ID INTEGER,
    Provider_Type TEXT,
    Location TEXT,
    Food_Type TEXT,
    Meal_Type TEXT,
    FOREIGN KEY (Provider_ID) REFERENCES Providers(Provider_ID)
);

CREATE TABLE IF NOT EXISTS Claims (
    Claim_ID INTEGER PRIMARY KEY,
    Food_ID INTEGER,
    Receiver_ID INTEGER,
    Status TEXT,
    Timestamp DATETIME,
    FOREIGN KEY (Food_ID) REFERENCES Food_Listings(Food_ID),
    FOREIGN KEY (Receiver_ID) REFERENCES Receivers(Receiver_ID)
);
"""

cursor.execute(schema_sql)
conn.commit()
```

1) How many food providers and receivers are there in each city?

```python
query_providers = """
    SELECT City, COUNT(*) AS Provider_Count
    FROM Providers
    GROUP BY City;
"""
cursor.execute(query_providers)
providers_result = cursor.fetchall()
providers_df = pd.DataFrame(providers_result, columns=["City", "Provider_Count"])

query_receivers = """
    SELECT City, COUNT(*) AS Receiver_Count
    FROM Receivers
    GROUP BY City;
"""
cursor.execute(query_receivers)
receivers_result = cursor.fetchall()
receivers_df = pd.DataFrame(receivers_result, columns=["City", "Receiver_Count"])

city_summary_df = pd.merge(providers_df, receivers_df, on="City", how="inner")

# Convert to int safely (no NaNs will be present)
city_summary_df["Provider_Count"] = city_summary_df["Provider_Count"].astype(int)
city_summary_df["Receiver_Count"] = city_summary_df["Receiver_Count"].astype(int)

print(city_summary_df)
```

2) Which type of food provider (restaurant, grocery store, etc.) contributes the most food?

```python
query = """
    SELECT Provider_Type, SUM(Quantity) AS Total_Quantity
    FROM Food_Listings 
    GROUP BY Provider_Type
    ORDER BY Total_Quantity DESC;
"""
cursor.execute(query)
result = cursor.fetchall()

df = pd.DataFrame(result, columns = ["Provider_Type","Total_Quantity"])
df
```

3) What is the contact information of food providers in a specific city?

```python
city_name = "New Jessica"

query = f"""
    SELECT Name, Contact, Address
    FROM Providers
    WHERE City = '{city_name}';
"""
cursor.execute(query)
result = cursor.fetchall()

df = pd.DataFrame(result, columns = ["Name","Contact","Address"])
df
```

4) Which receivers have claimed the most food?

```python
query = """
    SELECT r.Name AS Receiver_Name, SUM(f.Quantity) AS Total_Quantity_Claimed
    FROM Claims c
    JOIN Food_Listings f ON c.Food_ID = f.Food_ID
    JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID
    WHERE c.Status = 'Completed'  
    GROUP BY r.Receiver_ID, r.Name
    ORDER BY Total_Quantity_Claimed DESC
    LIMIT 10; 
"""
cursor.execute(query)
result = cursor.fetchall()

df = pd.DataFrame(result, columns = ["Receiver_Name","Total_Quantity_Claimed"])
df
```

5) What is the total quantity of food available from all providers?

```python
query = """
    SELECT Food_Name, SUM(Quantity) AS Total_Food_Quantity
    FROM Food_Listings
    GROUP BY Food_Name
    ORDER BY Total_Food_Quantity DESC;
"""
cursor.execute(query)
result = cursor.fetchall()

df = pd.DataFrame(result, columns = ["Food_Name","Total_Food_Quantity"])
df
```

6) Which city has the highest number of food listings?

```python
query = """
    SELECT Location AS City, COUNT(*) AS Food_Listings_Count
    FROM Food_Listings
    GROUP BY Location
    ORDER BY Food_Listings_Count DESC
    LIMIT 1;
"""

cursor.execute(query)
result = cursor.fetchall()

# Convert the result to a DataFrame for readability
df = pd.DataFrame(result, columns=["City", "Food_Listings_Count"])
print(df)
```

7) What are the most commonly available food types?

```python
query = """
    SELECT Food_Type, COUNT(*) AS Count
    FROM Food_Listings
    GROUP BY Food_Type
    ORDER BY Count DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# Convert the result to a DataFrame
df = pd.DataFrame(result, columns=["Food_Type", "Count"])
print(df)
```

8) How many food claims have been made for each food item?

```python
query = """
    SELECT f.Food_Name, COUNT(c.Claim_ID) AS Total_Claims
    FROM Claims c
    JOIN Food_Listings f ON c.Food_ID = f.Food_ID
    GROUP BY f.Food_Name
    ORDER BY Total_Claims DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load result into a DataFrame
df = pd.DataFrame(result, columns=["Food_Name", "Total_Claims"])
print(df)
```

9) Which provider has had the highest number of successful food claims?

```python
query = """
    SELECT p.Name AS Provider_Name, COUNT(c.Claim_ID) AS Successful_Claims
    FROM Claims c
    JOIN Food_Listings f ON c.Food_ID = f.Food_ID
    JOIN Providers p ON f.Provider_ID = p.Provider_ID
    WHERE c.Status = 'Completed'
    GROUP BY p.Provider_ID, p.Name
    ORDER BY Successful_Claims DESC
    LIMIT 1;
"""

cursor.execute(query)
result = cursor.fetchall()

# Convert result to DataFrame
df = pd.DataFrame(result, columns=["Provider_Name", "Successful_Claims"])
print(df)
```

10) What percentage of food claims are completed vs. pending vs. canceled?

```python
query = """
    SELECT Status, COUNT(*) AS Count
    FROM Claims
    GROUP BY Status;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["Status", "Count"])

# Calculate percentage
total_claims = df["Count"].sum()
df["Percentage"] = (df["Count"] / total_claims) * 100

print(df)
```

11) What is the average quantity of food claimed per receiver?

```python
query = """
    SELECT r.Name AS Receiver_Name, AVG(f.Quantity) AS Avg_Quantity_Claimed
    FROM Claims c
    JOIN Food_Listings f ON c.Food_ID = f.Food_ID
    JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID
    WHERE c.Status = 'Completed' 
    GROUP BY r.Receiver_ID, r.Name
    ORDER BY Avg_Quantity_Claimed DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["Receiver_Name", "Avg_Quantity_Claimed"])
print(df)
```

12) Which meal type (breakfast, lunch, dinner, snacks) is claimed the most?

```python
query = """
    SELECT f.Meal_Type, COUNT(c.Claim_ID) AS Claims_Count
    FROM Claims c
    JOIN Food_Listings f ON c.Food_ID = f.Food_ID
    WHERE c.Status = 'Completed'  -- consider only completed claims
    GROUP BY f.Meal_Type
    ORDER BY Claims_Count DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# Convert to DataFrame
df = pd.DataFrame(result, columns=["Meal_Type", "Claims_Count"])
print(df)
```

13) What is the total quantity of food donated by each provider?

```python
query = """
    SELECT p.Name AS Provider_Name, SUM(f.Quantity) AS Total_Quantity_Donated
    FROM Food_Listings f
    JOIN Providers p ON f.Provider_ID = p.Provider_ID
    GROUP BY p.Provider_ID, p.Name
    ORDER BY Total_Quantity_Donated DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["Provider_Name", "Total_Quantity_Donated"])
print(df)
```

```python
%%writefile foods_app.py
import streamlit as st
import pandas as pd
import psycopg2
from psycopg2 import sql
import plotly.express as px

# Database connection setup (update with your credentials)
def get_connection():
    return psycopg2.connect(
        host="localhost",
        user="postgres",
        password="Hari",
        dbname="food_data"
    )

# Utility: Run a read query and return DataFrame
def run_query(query):
    conn = get_connection()
    cur = conn.cursor()
    cur.execute(query)
    rows = cur.fetchall()
    colnames = [desc[0] for desc in cur.description]
    df = pd.DataFrame(rows, columns=colnames)
    conn.close()
    return df

# Load data for filtering
def load_options():
    conn = get_connection()
    food_df = pd.read_sql("SELECT DISTINCT Location, Food_Type, Provider_ID FROM Food_Listings", conn)
    providers_df = pd.read_sql("SELECT DISTINCT Provider_ID, Name FROM Providers", conn)
    conn.close()

    locs = food_df["Location"].dropna().unique().tolist() if "Location" in food_df.columns else []
    food_types = food_df["Food_Type"].dropna().unique().tolist() if "Food_Type" in food_df.columns else []
    providers = providers_df["Name"].dropna().unique().tolist() if "Name" in providers_df.columns else []

    return locs, providers, food_types

st.title("Local Food Wastage Management System")

menu = st.sidebar.radio("Menu", [
    "Home","Database Table", "Filter Donations", "Contact Directory", "CRUD Operations", "SQL Analytics", "SQL Queries & Visualization"
])

if menu == "Home":
    st.markdown("""
        This platform helps manage food donations, reduce waste, and connect providers with those in need.
        1) ***Providers:*** Restaurants, Households and Businesses list surplus food.
        2) ***Receivers:*** NGO's and Individuals claim available food.
        3) ***SQL Analysis:*** Powwerful insights using SQL queries.
    """)

elif menu == "Database Table":
    table = st.selectbox("Select Table",["Providers","Receivers","Food Listings"])
    if table == "Providers":
        df = run_query("SELECT * FROM Providers")
        st.dataframe(df)
    elif table == "Receivers":
        df = run_query("SELECT * FROM Receivers")
        st.dataframe(df)
    elif table == "Food Listings":
        df = run_query("SELECT * FROM Food_Listings")
        st.dataframe(df)

elif menu == "Filter Donations":
    st.subheader("Filter Food Donations")
    locs, providers, food_types = load_options()

    st.write("Locations:", locs)
    st.write("Providers:", providers)
    st.write("Food Types:", food_types)

    if locs and providers and food_types:
        location = st.selectbox("Select Location", locs)
        provider = st.selectbox("Select Provider", providers)
        ftype = st.selectbox("Select Food Type", food_types)

        query = f"""
            SELECT f.Food_Name, f.Quantity, f.Expiry_Date, p.Name AS Provider, f.Food_Type, f.Meal_Type, f.Location
            FROM Food_Listings f
            JOIN Providers p ON f.Provider_ID = p.Provider_ID
            WHERE f.Location = '{location}' AND p.Name = '{provider}' AND f.Food_Type = '{ftype}'
        """
        result_df = run_query(query)
        st.dataframe(result_df)
    else:
        st.warning("No data available to filter. Please check your database.")

elif menu == "Contact Directory":
    st.subheader("Contact Providers & Receivers")
    tab1, tab2 = st.tabs(["Providers", "Receivers"])

    with tab1:
        st.write("### Provider Contacts")
        df = run_query("SELECT Name, Type, City, Contact FROM Providers")
        st.dataframe(df)

    with tab2:
        st.write("### Receiver Contacts")
        df = run_query("SELECT Name, Type, City, Contact FROM Receivers")
        st.dataframe(df)

elif menu == "CRUD Operations":
    st.subheader("🛠️ Manage Food Listings")
    operation = st.radio("Choose Operation", ["Add", "Update", "Delete"])

    if operation == "Add":
        with st.form("Add Listing"):
            food_id = st.number_input("Food ID", min_value=1)
            name = st.text_input("Food Name")
            qty = st.number_input("Quantity", min_value=1)
            expiry = st.date_input("Expiry Date")
            pid = st.number_input("Provider ID", min_value=1)
            ptype = st.text_input("Provider Type")
            loc = st.text_input("Location")
            ftype = st.selectbox("Food Type", ["Vegetarian", "Non-Vegetarian", "Vegan"])
            mtype = st.selectbox("Meal Type", ["Breakfast", "Lunch", "Dinner", "Snacks"])
            submitted = st.form_submit_button("Add Listing")
            if submitted:
                conn = get_connection()
                cur = conn.cursor()
                cur.execute("SELECT 1 FROM Food_Listings WHERE Food_ID = %s", (food_id,))
                if cur.fetchone():
                    st.error("Food_ID already exists. Please choose a different ID.")
                else:
                    cur.execute("""
                        INSERT INTO Food_Listings VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
                    """, (food_id, name, qty, expiry, pid, ptype, loc, ftype, mtype))
                    conn.commit()
                    st.success("Listing added successfully!")
                conn.close()

    elif operation == "Update":
        with st.form("Update Listing"):
            fid = st.number_input("Food ID", min_value=1)
            qty = st.number_input("New Quantity", min_value=1)
            submitted = st.form_submit_button("Update Listing")
            if submitted:
                conn = get_connection()
                cur = conn.cursor()
                cur.execute("UPDATE Food_Listings SET Quantity = %s WHERE Food_ID = %s", (qty, fid))
                conn.commit()
                conn.close()
                st.success("Listing updated successfully!")

    elif operation == "Delete":
        fid = st.number_input("Food ID to Delete", min_value=1)
        if st.button("Delete Listing"):
            conn = get_connection()
            cur = conn.cursor()
            cur.execute("DELETE FROM Food_Listings WHERE Food_ID = %s", (fid,))
            conn.commit()
            conn.close()
            st.success("Deleted successfully!")

elif menu == "SQL Analytics":
    st.subheader("SQL Query Insights")
    queries = {
        "Providers per City": "SELECT City, COUNT(*) AS Provider_Count FROM Providers GROUP BY City;",
        "Receivers per City": "SELECT City, COUNT(*) AS Receiver_Count FROM Receivers GROUP BY City;",
        "Top Provider Types by Quantity": "SELECT Provider_Type, SUM(Quantity) AS Total FROM Food_Listings GROUP BY Provider_Type ORDER BY Total DESC;",
        "Most Claimed Meal Type": "SELECT Meal_Type, COUNT(*) AS Claims FROM Food_Listings f JOIN Claims c ON f.Food_ID = c.Food_ID WHERE c.Status = 'Completed' GROUP BY Meal_Type ORDER BY Claims DESC;",
        "Claims Status Distribution": "SELECT Status, COUNT(*) AS Count FROM Claims GROUP BY Status;",
        "Receivers Claimed Most Food": "SELECT r.Name AS Receiver_Name, SUM(f.Quantity) AS Total_Quantity FROM Claims c JOIN Food_Listings f ON c.Food_ID = f.Food_ID JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID WHERE c.Status = 'Completed' GROUP BY r.Receiver_ID, r.Name ORDER BY Total_Quantity DESC LIMIT 5;",
        "Food Listings per City": "SELECT Location, COUNT(*) AS Total_Listings FROM Food_Listings GROUP BY Location ORDER BY Total_Listings DESC;",
        "Common Food Types": "SELECT Food_Type, COUNT(*) AS Count FROM Food_Listings GROUP BY Food_Type ORDER BY Count DESC;",
        "Claims per Food Item": "SELECT Food_ID, COUNT(*) AS Total_Claims FROM Claims GROUP BY Food_ID ORDER BY Total_Claims DESC;",
        "Top Providers by Claims": "SELECT p.Name, COUNT(*) AS Successful_Claims FROM Providers p JOIN Food_Listings f ON p.Provider_ID = f.Provider_ID JOIN Claims c ON f.Food_ID = c.Food_ID WHERE c.Status = 'Completed' GROUP BY p.Name ORDER BY Successful_Claims DESC LIMIT 5;",
        "Food Claim Percentages": "SELECT Status, ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM Claims), 2) AS Percentage FROM Claims GROUP BY Status;",
        "Average Quantity per Receiver": "SELECT r.Name, ROUND(AVG(f.Quantity), 2) AS Avg_Quantity FROM Claims c JOIN Food_Listings f ON c.Food_ID = f.Food_ID JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID WHERE c.Status = 'Completed' GROUP BY r.Name ORDER BY Avg_Quantity DESC;",
        "Total Quantity by Provider": "SELECT p.Name, SUM(f.Quantity) AS Total_Quantity FROM Providers p JOIN Food_Listings f ON p.Provider_ID = f.Provider_ID GROUP BY p.Name ORDER BY Total_Quantity DESC;",
        "Meal Type by Quantity": "SELECT Meal_Type, SUM(Quantity) AS Total_Quantity FROM Food_Listings GROUP BY Meal_Type ORDER BY Total_Quantity DESC;",
        "City Food Stats": "SELECT f.Location AS City, COUNT(DISTINCT f.Provider_ID) AS Provider_Count, COUNT(DISTINCT r.Receiver_ID) AS Receiver_Count FROM Food_Listings f JOIN Claims c ON f.Food_ID = c.Food_ID JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID GROUP BY f.Location ORDER BY Provider_Count DESC;"
    }
    choice = st.selectbox("Choose Query", list(queries.keys()))
    df = run_query(queries[choice])
    st.dataframe(df)

elif menu == "SQL Queries & Visualization":
    import plotly.express as px

    st.subheader("SQL Queries & Visualization")
    queries = {
        "Total Food Listings": "SELECT COUNT(*) AS Total_Food_Listings FROM Food_Listings;",
        "Top 5 Providers by Quantity": """
            SELECT p.Name, SUM(f.Quantity) AS Total_Quantity
            FROM Providers p
            JOIN Food_Listings f ON p.Provider_ID = f.Provider_ID
            GROUP BY p.Name
            ORDER BY Total_Quantity DESC
            LIMIT 5;
        """
    }

    choice = st.selectbox("Choose Query", list(queries.keys()))
    df = run_query(queries[choice])
    st.dataframe(df)

    # Visualization
    if choice == "Top 5 Providers by Quantity":
        df.columns = df.columns.str.lower()  # Ensure all columns are lowercase
        fig = px.bar(df, x="name", y="total_quantity", color="name", title="Top 5 Providers by Quantity")
        st.plotly_chart(fig)

    elif choice == "Total Food Listings":
        st.metric("Total Food Listings", int(df.iloc[0, 0]))
```
