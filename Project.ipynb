```python
schema_sql = """
CREATE TABLE IF NOT EXISTS Providers (
    Provider_ID INTEGER PRIMARY KEY,
    Name TEXT,
    Type TEXT,
    Address TEXT,
    City TEXT,
    Contact TEXT
);

CREATE TABLE IF NOT EXISTS Receivers (
    Receiver_ID INTEGER PRIMARY KEY,
    Name TEXT,
    Type TEXT,
    City TEXT,
    Contact TEXT
);

CREATE TABLE IF NOT EXISTS Food_Listings (
    Food_ID INTEGER PRIMARY KEY,
    Food_Name TEXT,
    Quantity INTEGER,
    Expiry_Date DATE,
    Provider_ID INTEGER,
    Provider_Type TEXT,
    Location TEXT,
    Food_Type TEXT,
    Meal_Type TEXT,
    FOREIGN KEY (Provider_ID) REFERENCES Providers(Provider_ID)
);

CREATE TABLE IF NOT EXISTS Claims (
    Claim_ID INTEGER PRIMARY KEY,
    Food_ID INTEGER,
    Receiver_ID INTEGER,
    Status TEXT,
    Timestamp DATETIME,
    FOREIGN KEY (Food_ID) REFERENCES Food_Listings(Food_ID),
    FOREIGN KEY (Receiver_ID) REFERENCES Receivers(Receiver_ID)
);
"""

cursor.execute(schema_sql)
conn.commit()
```

3. What is the distribution of food expiry dates?

```python
query = """
SELECT
    CASE
        WHEN Expiry_Date < CURRENT_DATE THEN 'Expired'
        WHEN Expiry_Date BETWEEN CURRENT_DATE AND CURRENT_DATE + INTERVAL '7 days' THEN 'Expiring Soon'
        ELSE 'Fresh'
    END AS Expiry_Status,
    COUNT(*) AS Number_of_Items
FROM Food_Listings
GROUP BY Expiry_Status
ORDER BY Number_of_Items DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["Expiry_Date","Number_of_Items"])
print(df)
```

4. What is the claim rate by food type (claimed listings / total listings)?

```python
query = """
SELECT
    fl.Food_Type,
    COUNT(DISTINCT fl.Food_ID) AS Total_Listings,
    COUNT(DISTINCT c.Food_ID) AS Claimed_Listings,
    ROUND(COUNT(DISTINCT c.Food_ID)::decimal / NULLIF(COUNT(DISTINCT fl.Food_ID), 0) * 100, 2) AS Claim_Rate_Percent
FROM Food_Listings fl
LEFT JOIN Claims c ON fl.Food_ID = c.Food_ID
GROUP BY fl.Food_Type
ORDER BY Claim_Rate_Percent DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["Food_Type","Total_Listings","Claimed_Listings","Claim_Rate_Percent"])
print(df)
```

5. Which receivers have the most diversity in food types claimed?

7. What percentage of food listings were never claimed?

```python
query = """
SELECT
    ROUND(
        100.0 * (
            SELECT COUNT(*) FROM Food_Listings f
            WHERE f.Food_ID NOT IN (SELECT DISTINCT Food_ID FROM Claims)
        ) / (SELECT COUNT(*) FROM Food_Listings),
        2
    ) AS Percent_Unclaimed
;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["Percent_Unclaimed"])
print(df)
```

8. Which cities have the lowest number of receivers relative to providers?

```python
query = """
SELECT
    COALESCE(p.City, r.City) AS City,
    COUNT(DISTINCT p.Provider_ID) AS Provider_Count,
    COUNT(DISTINCT r.Receiver_ID) AS Receiver_Count,
    ROUND(
        CAST(COUNT(DISTINCT r.Receiver_ID) AS numeric) /
        NULLIF(COUNT(DISTINCT p.Provider_ID), 0), 2
    ) AS Receiver_to_Provider_Ratio
FROM
    Providers p
FULL OUTER JOIN Receivers r ON p.City = r.City
GROUP BY COALESCE(p.City, r.City)
ORDER BY Receiver_to_Provider_Ratio ASC NULLS LAST;
"""
cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["City","Provider_Count","Receiver_Count","Receiver_to_Provider_Ratio"])
print(df)
```

10. What is the distribution of claims by meal type over time?

```python
query = """
SELECT
    DATE(c.Timestamp) AS Claim_Date,
    f.Meal_Type,
    COUNT(c.Claim_ID) AS Total_Claims
FROM Claims c
JOIN Food_Listings f ON c.Food_ID = f.Food_ID
GROUP BY DATE(c.Timestamp), f.Meal_Type
ORDER BY Claim_Date, f.Meal_Type;
"""

cursor.execute(query)
result = cursor.fetchall()

# Load into DataFrame
df = pd.DataFrame(result, columns=["Claim_Date","Meal_Type","Total_Claims"])
print(df)
```

```python
%%writefile foods_app.py
import streamlit as st
import pandas as pd
import psycopg2
from psycopg2 import sql
import plotly.express as px

# Database connection setup
def get_connection():
    return psycopg2.connect(
        host="localhost",
        user="postgres",
        password="Hari",
        dbname="food_data"
    )

# Run a read query and return DataFrame
def run_query(query):
    conn = get_connection()
    cur = conn.cursor()
    cur.execute(query)
    rows = cur.fetchall()
    colnames = [desc[0] for desc in cur.description]
    df = pd.DataFrame(rows, columns=colnames)
    conn.close()
    return df

# Load data for filtering
def load_options():
    conn = get_connection()
    food_df = pd.read_sql("SELECT DISTINCT Location, Food_Type, Provider_ID FROM Food_Listings", conn)
    providers_df = pd.read_sql("SELECT DISTINCT Provider_ID, Name FROM Providers", conn)
    conn.close()

    locs = food_df["Location"].dropna().unique().tolist() if "Location" in food_df.columns else []
    food_types = food_df["Food_Type"].dropna().unique().tolist() if "Food_Type" in food_df.columns else []
    providers = providers_df["Name"].dropna().unique().tolist() if "Name" in providers_df.columns else []

    return locs, providers, food_types

st.title("Local Food Wastage Management System")

menu = st.sidebar.radio("Menu", [
    "Home","Database Table", "Filter Donations", "Contact Directory", "CRUD Operations", "SQL Analytics", "SQL Queries & Visualization"
])

if menu == "Home":
    st.markdown("""
        This platform helps manage food donations, reduce waste, and connect providers with those in need.
        1) ***Providers:*** Restaurants, Households and Businesses list surplus food.
        2) ***Receivers:*** NGO's and Individuals claim available food.
        3) ***SQL Analysis:*** Powerful insights using SQL queries.
    """)

elif menu == "Database Table":
    table = st.selectbox("Select Table",["Providers","Receivers","Food Listings","Claims"])
    if table == "Providers":
        df = run_query("SELECT * FROM Providers")
        st.dataframe(df)
    elif table == "Receivers":
        df = run_query("SELECT * FROM Receivers")
        st.dataframe(df)
    elif table == "Food Listings":
        df = run_query("SELECT * FROM Food_Listings")
        st.dataframe(df)
    elif table == "Claims":
        df = run_query("SELECT * FROM Claims")
        st.dataframe(df)

elif menu == "Filter Donations":
    st.subheader(" Filter Data")

    filter_option = st.selectbox("Select Data Type to Filter", ["Providers", "Receivers", "Claims", "Food Listings"])

    conn = get_connection()

    if filter_option == "Providers":
        st.markdown("### Filter Providers")
        provider_cols = pd.read_sql("SELECT * FROM Providers LIMIT 1", conn).columns.tolist()
        
        city_col = "City" if "City" in provider_cols else "city"
        type_col = "Type" if "Type" in provider_cols else "type"

        cities = pd.read_sql(f"SELECT DISTINCT {city_col} FROM Providers", conn)[city_col].dropna().tolist()
        types = pd.read_sql(f"SELECT DISTINCT {type_col} FROM Providers", conn)[type_col].dropna().tolist()

        selected_city = st.selectbox("Select City", cities)
        selected_type = st.selectbox("Select Provider Type", types)

        query = f"SELECT * FROM Providers WHERE {city_col} = %s AND {type_col} = %s"
        df = pd.read_sql(query, conn, params=(selected_city, selected_type))
        st.dataframe(df)

    elif filter_option == "Receivers":
        st.markdown("### Filter Receivers")
        receiver_cols = pd.read_sql("SELECT * FROM Receivers LIMIT 1", conn).columns.tolist()
        st.write("Receiver Columns:", receiver_cols)

        city_col = "City" if "City" in receiver_cols else "city"
        type_col = "Type" if "Type" in receiver_cols else "type"

        cities = pd.read_sql(f"SELECT DISTINCT {city_col} FROM Receivers", conn)[city_col].dropna().tolist()
        types = pd.read_sql(f"SELECT DISTINCT {type_col} FROM Receivers", conn)[type_col].dropna().tolist()

        selected_city = st.selectbox("Select City", cities)
        selected_type = st.selectbox("Select Receiver Type", types)

        query = f"SELECT * FROM Receivers WHERE {city_col} = %s AND {type_col} = %s"
        df = pd.read_sql(query, conn, params=(selected_city, selected_type))
        st.dataframe(df)

    elif filter_option == "Claims":
        st.markdown("### Filter Claims by Status")
        claim_cols = pd.read_sql("SELECT * FROM Claims LIMIT 1", conn).columns.str.lower().tolist()
        status_col = "status" if "status" in claim_cols else "Status"

        statuses = pd.read_sql(f"SELECT DISTINCT {status_col} FROM Claims", conn)[status_col].dropna().tolist()
        selected_status = st.selectbox("Select Claim Status", statuses)

        query = f"""
            SELECT c.Claim_ID, f.Food_Name, p.Name AS Provider, r.Name AS Receiver, c.{status_col}
            FROM Claims c
            JOIN Food_Listings f ON c.Food_ID = f.Food_ID
            JOIN Providers p ON f.Provider_ID = p.Provider_ID
            JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID
            WHERE c.{status_col} = %s
        """
        df = pd.read_sql(query, conn, params=(selected_status,))
        st.dataframe(df)

    elif filter_option == "Food Listings":
        st.markdown("### Filter Food Listings")
        food_df = pd.read_sql("SELECT * FROM Food_Listings", conn)
        food_cols = [col.lower() for col in food_df.columns]

        location_col = next((col for col in food_df.columns if col.lower() == "location"), None)
        food_type_col = next((col for col in food_df.columns if col.lower() == "food_type"), None)

        provider_df = pd.read_sql("SELECT * FROM Providers", conn)
        provider_type_col = next((col for col in provider_df.columns if col.lower() == "type"), None)

        if location_col and food_type_col and provider_type_col:
            locations = food_df[location_col].dropna().unique().tolist()
            food_types = food_df[food_type_col].dropna().unique().tolist()
            provider_types = provider_df[provider_type_col].dropna().unique().tolist()

            selected_location = st.selectbox("Select Location", locations)
            selected_ftype = st.selectbox("Select Food Type", food_types)
            selected_ptype = st.selectbox("Select Provider Type", provider_types)

            query = f"""
                SELECT f.Food_Name, f.Quantity, f.Expiry_Date, 
                       f.{food_type_col} AS Food_Type, f.{location_col} AS Location,
                       p.{provider_type_col} AS Provider_Type, p.Name AS Provider_Name
                FROM Food_Listings f
                JOIN Providers p ON f.Provider_ID = p.Provider_ID
                WHERE f.{location_col} = %s AND f.{food_type_col} = %s AND p.{provider_type_col} = %s
            """
            df = pd.read_sql(query, conn, params=(selected_location, selected_ftype, selected_ptype))
            st.dataframe(df)
        else:
            st.error("Required columns 'Location', 'Food_Type' or 'Type' not found in Food Listings or Providers table.")

elif menu == "Contact Directory":
    st.subheader("Contact Providers & Receivers")
    tab1, tab2 = st.tabs(["Providers", "Receivers"])

    with tab1:
        st.write("### Get Provider Contact by ID")
        pid = st.number_input("Enter Provider ID", min_value=1, step=1)
        if st.button("Get Provider Details"):
            query = f"SELECT * FROM Providers WHERE Provider_ID = {pid};"
            df = run_query(query)
            if df.empty:
                st.error("No provider found with that ID.")
            else:
                st.success("Provider Details Retrieved:")
                st.dataframe(df)

    with tab2:
        st.write("### Get Receiver Contact by ID")
        rid = st.number_input("Enter Receiver ID", min_value=1, step=1, key="receiver_id")
        if st.button("Get Receiver Details"):
            query = f"SELECT * FROM Receivers WHERE Receiver_ID = {rid};"
            df = run_query(query)
            if df.empty:
                st.error("No receiver found with that ID.")
            else:
                st.success("Receiver Details Retrieved:")
                st.dataframe(df)


elif menu == "CRUD Operations":
    st.subheader("Manage Database Records")

    table_option = st.selectbox("Select Table", ["Providers", "Receivers", "Food Listings", "Claims"])
    operation = st.radio("Choose Operation", ["Add", "Update", "Delete"])

    conn = get_connection()
    cur = conn.cursor()

    # Providers
    if table_option == "Providers":
        if operation == "Add" or operation == "Update":
            with st.form("Provider Form"):
                pid = st.number_input("Provider ID", min_value=1)
                name = st.text_input("Name")
                ptype = st.text_input("Type")
                address = st.text_input("Address")
                city = st.text_input("City")
                contact = st.text_input("Contact")
                submitted = st.form_submit_button(f"{operation} Provider")
                if submitted:
                    if operation == "Add":
                        cur.execute("SELECT 1 FROM Providers WHERE Provider_ID = %s", (pid,))
                        if cur.fetchone():
                            st.error("Provider ID already exists.")
                        else:
                            cur.execute("""
                                INSERT INTO Providers (Provider_ID, Name, Type, Address, City, Contact)
                                VALUES (%s, %s, %s, %s, %s, %s)
                            """, (pid, name, ptype, address, city, contact))
                            conn.commit()
                            st.success("Provider added!")
                    else:
                        cur.execute("SELECT 1 FROM Providers WHERE Provider_ID = %s", (pid,))
                        if cur.fetchone():
                            cur.execute("""
                                UPDATE Providers SET Name=%s, Type=%s, Address=%s, City=%s, Contact=%s
                                WHERE Provider_ID = %s
                            """, (name, ptype, address, city, contact, pid))
                            conn.commit()
                            st.success("Provider updated!")
                        else:
                            st.warning("Provider not found.")

        elif operation == "Delete":
            pid = st.number_input("Provider ID to Delete", min_value=1)
            if st.button("Delete Provider"):
                cur.execute("DELETE FROM Providers WHERE Provider_ID = %s", (pid,))
                conn.commit()
                st.success("Provider deleted!")

    # Receivers
    elif table_option == "Receivers":
        if operation == "Add" or operation == "Update":
            with st.form("Receiver Form"):
                rid = st.number_input("Receiver ID", min_value=1)
                name = st.text_input("Name")
                rtype = st.text_input("Type")
                city = st.text_input("City")
                contact = st.text_input("Contact")
                submitted = st.form_submit_button(f"{operation} Receiver")
                if submitted:
                    if operation == "Add":
                        cur.execute("SELECT 1 FROM Receivers WHERE Receiver_ID = %s", (rid,))
                        if cur.fetchone():
                            st.error("Receiver ID already exists.")
                        else:
                            cur.execute("""
                                INSERT INTO Receivers (Receiver_ID, Name, Type, City, Contact)
                                VALUES (%s, %s, %s, %s, %s)
                            """, (rid, name, rtype, city, contact))
                            conn.commit()
                            st.success("Receiver added!")
                    else:
                        cur.execute("SELECT 1 FROM Receivers WHERE Receiver_ID = %s", (rid,))
                        if cur.fetchone():
                            cur.execute("""
                                UPDATE Receivers SET Name=%s, Type=%s, City=%s, Contact=%s
                                WHERE Receiver_ID = %s
                            """, (name, rtype, city, contact, rid))
                            conn.commit()
                            st.success("Receiver updated!")
                        else:
                            st.warning("Receiver not found.")

        elif operation == "Delete":
            rid = st.number_input("Receiver ID to Delete", min_value=1)
            if st.button("Delete Receiver"):
                cur.execute("DELETE FROM Receivers WHERE Receiver_ID = %s", (rid,))
                conn.commit()
                st.success("Receiver deleted!")

    # Food Listings
    elif table_option == "Food Listings":
        if operation == "Add" or operation == "Update":
            with st.form("Food Listing Form"):
                fid = st.number_input("Food ID", min_value=1)
                name = st.text_input("Food Name")
                qty = st.number_input("Quantity", min_value=1)
                expiry = st.date_input("Expiry Date")
                pid = st.number_input("Provider ID", min_value=1)
                ptype = st.text_input("Provider Type")
                loc = st.text_input("Location")
                ftype = st.selectbox("Food Type", ["Vegetarian", "Non-Vegetarian", "Vegan"])
                mtype = st.selectbox("Meal Type", ["Breakfast", "Lunch", "Dinner", "Snacks"])
                submitted = st.form_submit_button(f"{operation} Listing")
                if submitted:
                    if operation == "Add":
                        cur.execute("SELECT 1 FROM Food_Listings WHERE Food_ID = %s", (fid,))
                        if cur.fetchone():
                            st.error("Food ID already exists.")
                        else:
                            cur.execute("""
                                INSERT INTO Food_Listings (Food_ID, Food_Name, Quantity, Expiry_Date,
                                Provider_ID, Provider_Type, Location, Food_Type, Meal_Type)
                                VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
                            """, (fid, name, qty, expiry, pid, ptype, loc, ftype, mtype))
                            conn.commit()
                            st.success("Listing added!")
                    else:
                        cur.execute("SELECT 1 FROM Food_Listings WHERE Food_ID = %s", (fid,))
                        if cur.fetchone():
                            cur.execute("""
                                UPDATE Food_Listings SET Food_Name=%s, Quantity=%s, Expiry_Date=%s,
                                Provider_ID=%s, Provider_Type=%s, Location=%s, Food_Type=%s, Meal_Type=%s
                                WHERE Food_ID = %s
                            """, (name, qty, expiry, pid, ptype, loc, ftype, mtype, fid))
                            conn.commit()
                            st.success("Listing updated!")
                        else:
                            st.warning("Food ID not found.")

        elif operation == "Delete":
            fid = st.number_input("Food ID to Delete", min_value=1)
            if st.button("Delete Listing"):
                cur.execute("DELETE FROM Food_Listings WHERE Food_ID = %s", (fid,))
                conn.commit()
                st.success("Listing deleted!")

    # Claims
    elif table_option == "Claims":
        if operation == "Add" or operation == "Update":
            with st.form("Claim Form"):
                cid = st.number_input("Claim ID", min_value=1)
                fid = st.number_input("Food ID", min_value=1)
                rid = st.number_input("Receiver ID", min_value=1)
                status = st.selectbox("Status", ["Pending", "Approved", "Rejected"])
                timestamp = st.date_input("Timestamp")
                submitted = st.form_submit_button(f"{operation} Claim")
                if submitted:
                    if operation == "Add":
                        cur.execute("SELECT 1 FROM Claims WHERE Claim_ID = %s", (cid,))
                        if cur.fetchone():
                            st.error("Claim ID already exists.")
                        else:
                            cur.execute("""
                                INSERT INTO Claims (Claim_ID, Food_ID, Receiver_ID, Status, Timestamp)
                                VALUES (%s, %s, %s, %s, %s)
                            """, (cid, fid, rid, status, timestamp))
                            conn.commit()
                            st.success("Claim added!")
                    else:
                        cur.execute("SELECT 1 FROM Claims WHERE Claim_ID = %s", (cid,))
                        if cur.fetchone():
                            cur.execute("""
                                UPDATE Claims SET Food_ID=%s, Receiver_ID=%s, Status=%s, Timestamp=%s
                                WHERE Claim_ID = %s
                            """, (fid, rid, status, timestamp, cid))
                            conn.commit()
                            st.success("Claim updated!")
                        else:
                            st.warning("Claim not found.")

        elif operation == "Delete":
            cid = st.number_input("Claim ID to Delete", min_value=1)
            if st.button("Delete Claim"):
                cur.execute("DELETE FROM Claims WHERE Claim_ID = %s", (cid,))
                conn.commit()
                st.success("🗑️ Claim deleted!")

    conn.close()

elif menu == "SQL Analytics":
    st.subheader("SQL Query Insights")
    queries = {      
        "1) Providers and Receivers per City": "SELECT COALESCE(p.City, r.City) AS city, COUNT(DISTINCT p.Provider_ID) AS provider_count, COUNT(DISTINCT r.Receiver_ID) AS receiver_count FROM Providers p FULL OUTER JOIN Receivers r ON p.City = r.City GROUP BY COALESCE(p.City, r.City) ORDER BY city;",
        "2) Top Provider Types by Quantity": "SELECT Provider_Type, SUM(Quantity) AS Total FROM Food_Listings GROUP BY Provider_Type ORDER BY Total DESC;",
        "3) Provider Contacts by City (New Jessica) ": "SELECT Name, Type, Address, Contact FROM Providers WHERE City = 'New Jessica';",
        "4) Receivers Claimed Most Food": "SELECT r.Name AS Receiver_Name, SUM(f.Quantity) AS Total_Quantity FROM Claims c JOIN Food_Listings f ON c.Food_ID = f.Food_ID JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID WHERE c.Status = 'Completed' GROUP BY r.Receiver_ID, r.Name ORDER BY Total_Quantity DESC LIMIT 5;",
        "5) Total Food Quantity Available": "SELECT Food_Name, SUM(Quantity) AS Total_Food_Quantity FROM Food_Listings GROUP BY Food_Name ORDER BY Total_Food_Quantity DESC;",
        "6) Highest number of food listings":"SELECT Location AS City, COUNT(*) AS Food_Listings_Count FROM Food_Listings GROUP BY Location ORDER BY Food_Listings_Count DESC LIMIT 1;",
        "7) Common Food Types": "SELECT Food_Type, COUNT(*) AS Count FROM Food_Listings GROUP BY Food_Type ORDER BY Count DESC;",
        "8) Food Claims for Each Food item":"SELECT f.Food_Name, COUNT(c.Claim_ID) AS Total_Claims FROM Claims c JOIN Food_Listings f ON c.Food_ID = f.Food_ID GROUP BY f.Food_Name ORDER BY Total_Claims DESC;",
        "9) Highest number of successful food claims":"SELECT p.Name AS Provider_Name, COUNT(c.Claim_ID) AS Successful_Claims FROM Claims c JOIN Food_Listings f ON c.Food_ID = f.Food_ID JOIN Providers p ON f.Provider_ID = p.Provider_ID WHERE c.Status = 'Completed' GROUP BY p.Provider_ID, p.Name ORDER BY Successful_Claims DESC LIMIT 1;",
        "10) Food Claim Percentages": "SELECT Status, ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM Claims), 2) AS Percentage FROM Claims GROUP BY Status;",
        "11) Average quantity of food claimed per receiver":"SELECT r.Name AS Receiver_Name, AVG(f.Quantity) AS Avg_Quantity_Claimed FROM Claims c JOIN Food_Listings f ON c.Food_ID = f.Food_ID JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID WHERE c.Status = 'Completed' GROUP BY r.Receiver_ID, r.Name ORDER BY Avg_Quantity_Claimed DESC;",
        "12) Meal Type by Quantity": "SELECT Meal_Type, SUM(Quantity) AS Total_Quantity FROM Food_Listings GROUP BY Meal_Type ORDER BY Total_Quantity DESC;",
        "13) Total quantity of food donated by each provider":"SELECT p.Name AS Provider_Name, SUM(f.Quantity) AS Total_Quantity_Donated FROM Food_Listings f JOIN Providers p ON f.Provider_ID = p.Provider_ID GROUP BY p.Provider_ID, p.Name ORDER BY Total_Quantity_Donated DESC;"
    }
    
    choice = st.selectbox("Choose Query", list(queries.keys()))
    df = run_query(queries[choice])
    st.dataframe(df)

elif menu == "SQL Queries & Visualization":
    import plotly.express as px

    st.subheader("SQL Queries & Visualization")

    queries = {
        "1) Top 5 Providers by Quantity": {
            "sql": """
                SELECT p.Name, SUM(f.Quantity) AS Total_Quantity
                FROM Providers p
                JOIN Food_Listings f ON p.Provider_ID = f.Provider_ID
                GROUP BY p.Name
                ORDER BY Total_Quantity DESC
                LIMIT 5;
            """,
            "visual": "bar",
            "x": "name",
            "y": "total_quantity",
            "title": "Top 5 Providers by Quantity"
        },
        "2) Food Donation listings by City": {
            "sql": """
                SELECT Location AS city,
                COUNT(*) AS total_listings
                FROM Food_Listings
                GROUP BY Location
                ORDER BY total_listings DESC
                LIMIT 10;
                """,
                "visual": "bar",
                "x": "city",
                "y": "total_listings",
                "title": "Top 10 Cities by Food Listings"
        },
        "3) Expiry Status Distribution": {
            "sql": """
                SELECT
                    CASE
                        WHEN Expiry_Date < CURRENT_DATE THEN 'Expired'
                        WHEN Expiry_Date BETWEEN CURRENT_DATE AND CURRENT_DATE + INTERVAL '7 days' THEN 'Expiring Soon'
                        ELSE 'Fresh'
                    END AS Expiry_Status,
                    COUNT(*) AS Number_of_Items
                FROM Food_Listings
                GROUP BY Expiry_Status
                ORDER BY Number_of_Items DESC;
            """,
            "visual": "none"
        },
        "4) Claim Rate by Food Type": {
            "sql": """
                SELECT
                    fl.Food_Type,
                    COUNT(DISTINCT fl.Food_ID) AS Total_Listings,
                    COUNT(DISTINCT c.Food_ID) AS Claimed_Listings,
                    ROUND(COUNT(DISTINCT c.Food_ID)::decimal / NULLIF(COUNT(DISTINCT fl.Food_ID), 0) * 100, 2) AS Claim_Rate_Percent
                FROM Food_Listings fl
                LEFT JOIN Claims c ON fl.Food_ID = c.Food_ID
                GROUP BY fl.Food_Type
                ORDER BY Claim_Rate_Percent DESC;
            """,
            "visual": "none"
        },
        "5) Top 5 Receivers by Food Type Diversity": {
            "sql": """
                SELECT 
                    r.Receiver_ID,
                    r.Name AS Receiver_Name,
                    COUNT(DISTINCT f.Food_Type) AS Unique_Food_Types_Claimed
                FROM Claims c
                JOIN Food_Listings f ON c.Food_ID = f.Food_ID
                JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID
                GROUP BY r.Receiver_ID, r.Name
                ORDER BY Unique_Food_Types_Claimed DESC
                LIMIT 5;
            """,
            "visual": "none"
        },
        "6) Claims by Receiver Type": {
            "sql": """
                SELECT
                    r.Type AS Receiver_Type,
                    COUNT(c.Claim_ID) AS Total_Claims
                FROM Claims c
                JOIN Receivers r ON c.Receiver_ID = r.Receiver_ID
                GROUP BY r.Type
                ORDER BY Total_Claims DESC;
            """,
            "visual": "none"
        },
        "7) Percentage of Unclaimed Food Listings": {
            "sql": """
                SELECT
                    ROUND(
                        100.0 * (
                            SELECT COUNT(*) FROM Food_Listings f
                            WHERE f.Food_ID NOT IN (SELECT DISTINCT Food_ID FROM Claims)
                        ) / (SELECT COUNT(*) FROM Food_Listings),
                        2
                    ) AS Percent_Unclaimed;
            """,
            "visual": "none"
        },
        "8) Receiver to Provider Ratio per City": {
            "sql": """
                SELECT
                    COALESCE(p.City, r.City) AS City,
                    COUNT(DISTINCT p.Provider_ID) AS Provider_Count,
                    COUNT(DISTINCT r.Receiver_ID) AS Receiver_Count,
                    ROUND(
                        CAST(COUNT(DISTINCT r.Receiver_ID) AS numeric) /
                        NULLIF(COUNT(DISTINCT p.Provider_ID), 0), 2
                    ) AS Receiver_to_Provider_Ratio
                FROM
                    Providers p
                FULL OUTER JOIN Receivers r ON p.City = r.City
                GROUP BY COALESCE(p.City, r.City)
                ORDER BY Receiver_to_Provider_Ratio ASC NULLS LAST;
            """,
            "visual": "none"
        },
        "9) Providers With No Donations": {
        "sql": """
            SELECT p.Provider_ID, p.Name, p.Type, p.City, p.Contact
            FROM Providers p
            LEFT JOIN Food_Listings f ON p.Provider_ID = f.Provider_ID
            WHERE f.Food_ID IS NULL;
        """,
        "visual": "none"
        },
        "10) Claims by Meal Type Over Time": {
        "sql": """
            SELECT
            DATE(c.Timestamp) AS Claim_Date,
            f.Meal_Type,
            COUNT(c.Claim_ID) AS Total_Claims
            FROM Claims c
            JOIN Food_Listings f ON c.Food_ID = f.Food_ID
            GROUP BY DATE(c.Timestamp), f.Meal_Type
            ORDER BY Claim_Date, f.Meal_Type;
        """,
        "visual": "line"
        }
    }

    choice = st.selectbox("Choose Query", list(queries.keys()))
    query = queries[choice]["sql"]
    df = run_query(query)
    st.dataframe(df)

    if queries[choice]["visual"] == "bar":
        df.columns = df.columns.str.lower()
        fig = px.bar(df, x=queries[choice]["x"], y=queries[choice]["y"], color=queries[choice]["x"], title=queries[choice]["title"])
        st.plotly_chart(fig)
```
